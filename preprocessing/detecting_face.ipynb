```python
import pandas as pd
import numpy as np
import os
from retinaface import RetinaFace
from tqdm import tqdm
import cv2
```

```python
import os

def find_folder(name, path):
    for root, dirs, files in os.walk(path):
        if name in dirs:
            return os.path.join(root, name)
    return None

# This searches your entire Drive for the folder named 'real'
found_path = find_folder('real', '/content/drive/MyDrive')

if found_path:
    print(f"üéØ FOUND IT! Use this exact path:")
    print(f"input_folder = '{found_path}'")
else:
    print("‚ùì I couldn't find a folder named 'real'.")
    print("Let's see what IS inside DeepfakeDetector:")
    print(os.listdir('/content/drive/MyDrive/DeepfakeDetector'))
```

```python
# root_input = "videoframes"
# root_output = "faceframes"

root_input = '/content/drive/MyDrive/DeepfakeDetector/DeepFakeDetectionSystem/videoframes'
root_output = '/content/drive/MyDrive/DeepfakeDetector/DeepFakeDetectionSystem/faceframes'

for label in ["real","fake"]:
    input_folder = os.path.join(root_input,label)
    output_folder = os.path.join(root_output,label)

    os.makedirs(output_folder,exist_ok=True)

    for videos in os.listdir(input_folder):
        input_video_path = os.path.join(input_folder,videos)
        output_video_path = os.path.join(output_folder,videos)

        os.makedirs(output_video_path,exist_ok=True)

        for frame_name in tqdm(os.listdir(input_video_path)):
            frame_path = os.path.join(input_video_path,frame_name)

            img = cv2.imread(frame_path)

            if img is None:
                continue
            
            detections = RetinaFace.detect_faces(img)

            if isinstance(detections,dict):

                first_face = list(detections.values())[0]
                x1,y1,x2,y2 = first_face["facial_area"]

                crop_face = img[y1:y2, x1:x2]
                crop_face = cv2.resize(crop_face,(224,224))

                save_path = os.path.join(output_video_path,frame_name)
                cv2.imwrite(save_path,crop_face)
```
